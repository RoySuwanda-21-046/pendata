## **1. K-NN**
KNN merupakan algoritma klasifikasi yang paling sederhana dalam mengklasifikasikan sebuah gambar kedalam sebuah label. Metode ini mudah dipahami dibandingkan metode lain karena mengklasifikasikan berdasarkan jarak terdekat dengan objek lain (tetangga).




## Mengambil Data Dari Dataset yang telah diberikan





```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv')
dataset
```

## Mengeksplorasi Data





```python
dataset.head()
```

## Memanggil fungsi ".shape" untuk mengetahui jumlah baris dan kolom dari dataset

```python
dataset.shape
```

## Memindahkan Dataset Kedalam Input (X) dan target (Y)


*   Membuat dataframe dengan semua training data kecuali kolom target
*   Mengecek apakah variabel target sudah dihapus







```python
X = dataset.drop(columns=["target"])
X.head()
```

```python
Y = dataset["target"].values
Y[0:303]
```

## Memisahkan dataset menjadi train set dan test set





```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1, stratify=Y)
```

## Scaling Data
Membuat numerical data pada dataset memiliki rentang nilai (scale) yang sama





```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Membuat dan Testing Model





```python
from sklearn.neighbors import KNeighborsClassifier
# Membuat KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_scaled,Y_train)
```

## Menampilkan 5 model prediksi pertama pada data test

```python
knn.predict(X_test_scaled)[0:5]
```

## Mengecek akurasi dari model yang dibuat dengan test  data

```python
knn.score(X_test_scaled, Y_test)
```

## Kesimpulan Dari Klasifikasi dengan K-NN



*    Data Heart Disease terdiri dari 303 baris dan 14 kolom
*   Nilai 5 prediksi pertama pada data test bernilai 1
*   Nilai akurasi dari perhitungan tadi sebesar 0,80327



## **2. Naive Bayes**
Naive bayes merupakan metode pengklasifikasian berdasarkan probabilitas sederhana dan dirancang agar dapat dipergunakan dengan asumsi antar variabel penjelas saling bebas (independen). Pada algoritma ini pembelajaran lebih ditekankan pada pengestimasian probabilitas.

Tahapan dari proses algoritma Naive Bayes adalah:
1. Menghitung jumlah kelas / label,
2. Menghitung Jumlah Kasus Per Kelas,
3. Kalikan Semua Variable Kelas,
4. Bandingkan Hasil Per Kelas.

## menentukan variabel independen dan variabel dependen dari data yang akan dianalisis. 

```python
# Variabel independen
x = dataset.drop(['target'], axis = 1)
x.head()
```

kolom target di drop atau dihapus dari data frame karena menjadi variabel dependen

```python
# Variabel dependen
y = dataset['target']
y.head()
```

## Klasifikasi Naive Bayes


## Melakukan *Train Test Split* yang akan membagi dataset menjadi training set dan test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

membagi dataset menjadi training set sebesar 80% dan test set sebesar 20%.

## Prediksi Training Set dan Test Set


*   Import Gaussian Naive Bayes model
*   Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
*    Memasukkan data training pada fungsi klasifikasi naive bayes





```python
from sklearn.naive_bayes import GaussianNB
modelnb = GaussianNB()
nbtrain = modelnb.fit(x_train, y_train)
nbtrain.class_count_
```

## Menentukan Hasil Prediksi

```python
y_pred = nbtrain.predict(x_test)
y_pred
```

## Menentukan probabilitas hasil prediksi

```python
nbtrain.predict_proba(x_test)
```

## Import confusion_matrix model

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

## Menghitung nilai akurasi dari klasifikasi naive bayes

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
