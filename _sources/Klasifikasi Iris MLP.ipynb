```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # visualization
from sklearn.neural_network import MLPClassifier # neural network
from sklearn.model_selection import train_test_split
from sklearn import metrics

```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Classification/master/IrisDataset.csv')
dataset.sample(5)
```

```python
dataset.head(5)
```

```python
df_norm = dataset[['sepal_length','sepal_width','petal_length','petal_width']].apply(lambda x: (x-x.min()) / (x.max() - x.min()))
df_norm.sample(n=5)
```

```python
df_norm.describe()
```

```python
target = dataset[['species']].replace(['setosa','versicolor','virginica'],[0,1,2])
target.sample(n=5)
```

```python
df = pd.concat([df_norm,target], axis=1)
df.sample(n=5)
```

```python
train, test = train_test_split(df, test_size = 0.3)
trainX = train[['sepal_length','sepal_width','petal_length','petal_width']]# taking the training data features
trainY=train.species# output of our training data
testX= test[['sepal_length','sepal_width','petal_length','petal_width']] # taking test data features
testY =test.species   #output value of test data
trainX.head(5)
```

```python
trainY.head(5)
```

```python
testX.head(5)
```

```python
testY.head(5)
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(3, 3), random_state=1)
```

```python
clf.fit(trainX, trainY)
```

```python
prediction = clf.predict(testX)
print(prediction)
```

```python
print(testY.values)
```

```python
print('The accuracy of the Multi-layer Perceptron is:',metrics.accuracy_score(prediction,testY))
```
